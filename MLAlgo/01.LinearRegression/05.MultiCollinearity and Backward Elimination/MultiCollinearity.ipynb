```python
import pandas as pd
```

```python
import statsmodels.api as sm
df_adv = pd.read_csv('Advertising.csv', index_col=0)
X = df_adv[['TV', 'radio','newspaper']]
y = df_adv['sales']
df_adv.head()
```

```python
X.corr()
```

```python
X = sm.add_constant(X)
X
```

```python
## fit a OLS model with intercept on TV and Radio

model= sm.OLS(y, X).fit()
```

```python
model.summary()
```

```python
import matplotlib.pyplot as plt
X.iloc[:,1:].corr()
```

```python
df_salary = pd.read_csv('Salary_Data.csv')
df_salary.head()
```

```python
X = df_salary[['YearsExperience', 'Age']]
y = df_salary['Salary']
```

```python
X.corr()
```

```python
## fit a OLS model with intercept on TV and Radio
X = sm.add_constant(X)
model= sm.OLS(y, X).fit()
```

```python
model.summary()
```

# From above we can see p value of age is high drop  that row and generate LinearRegression Model

```python
X = df_salary[['YearsExperience']]
X
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
reg.fit(X,y)
```

```python
reg.predict([[5]])
```

```python
reg.score(X,y)
```

```python
%matplotlib inline
plt.xlabel('Experience')
plt.ylabel('Salary')
plt.title('Salary Data')
plt.scatter(X,y,color='red',marker='+')
# predicted response vector 
y = reg.intercept_ + reg.coef_* X
# plotting the regression line 
plt.plot(X, y, color = "g") 
```
