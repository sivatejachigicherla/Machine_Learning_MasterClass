```python
# Simple Nyquist plotting

import control
import matplotlib.pyplot as plt

num = [0,1,2]
den = [1,0,-1]

#Creating a transfer function G = num/den
G = control.tf(num,den) 

control.nyquist(G)
plt.grid(True)
plt.xlabel('Re(s)')
plt.ylabel('Im(s)')
plt.show()
```
